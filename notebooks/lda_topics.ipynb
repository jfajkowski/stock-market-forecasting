```python
import gensim
import pyLDAvis.gensim

import pandas as pd

df = pd.read_csv('../data/interim/Corpus_Cleaned.csv')
text_data = df.loc[:, 'Top1':'Top25'].applymap(lambda x: str(x).split(' ')).values.flatten()
print('Loaded {} samples.'.format(text_data))

dictionary = gensim.corpora.Dictionary(text_data)
corpus = [dictionary.doc2bow(text) for text in text_data]
```

```python
NUM_TOPICS = 5
print('LDA model training started.')
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
print('LDA model training finished.')
for topic in ldamodel.print_topics():
    print(topic)
    
lda_display = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
```
